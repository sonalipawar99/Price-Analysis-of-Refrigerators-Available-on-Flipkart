```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd

# Initialize lists
title = []
rating_1 = []
Review_rating = []
Capacity = []
Discount = []
Price = []
compressor = []
Brands = []


for i in range(1, 21):  
    url = f"https://www.flipkart.com/search?q=refrigerator&page={i}"
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")

    for item in soup.find_all("div", class_='tUxRFH'):
        text = item.text

        # Title
        title_tag = item.find("div", class_='KzDlHZ')
        title.append(title_tag.text if title_tag else np.nan)

        # Rating
        rating_tag = item.find("div", class_='XQDdHH')
        rating_1.append(rating_tag.text if rating_tag else np.nan)

        # Review & Rating
        review_tag = item.find("div", class_='_5OesEi')
        Review_rating.append(review_tag.text if review_tag else np.nan)

        # Capacity (e.g., 190 L)
        ca = re.findall(r'(\d+)\s*L', text)
        Capacity.append(ca[0] if ca else np.nan)

        # Discount (e.g., 20% off)
        di = re.findall(r'(\d+)%\s*off', text)
        Discount.append(di[0] if di else np.nan)

        # Price
        price_tag = item.find("div", class_='Nx9bqj _4b5DiR')
        Price.append(price_tag.text if price_tag else np.nan)

        # Compressor Warranty
        co = re.findall(r'(\d+)\s+Years?\s+Warranty on Compressor', text)
        compressor.append(co[0] if co else np.nan)

        # Brand
        brand_tag = item.find("div", class_='fxf7w6 rgHxCQ')
        Brands.append(brand_tag.text.strip() if brand_tag else np.nan)

# Create DataFrame
dic = pd.DataFrame({
    "Title": title,
    "Rating": rating_1,
    "Review_rating": Review_rating,
    "Capacity": Capacity,
    "Discount": Discount,
    "Price": Price,
    "Compressor_Warranty_Years": compressor,
    "Brand": Brands
})
refrigerator_df = pd.DataFrame(dic)       #  create dataFrame:
df.to_csv("refrigerators.csv", index=False)
print(df.head())
```

## Incorrect Format of the data (Invalid Values, Columns)

```python
# Remove ₹ and , and convert to float
df['price'] = df['price'].str.replace('₹', '').str.replace(',', '').astype(float)

# Remove '%' from discount and convert to float
df['discount'] = df['discount'].str.replace('%', '').astype(float)
```

## Converting the data types 

```python
# Check missing values
print(df.isnull().sum())

# Fill missing ratings/reviews with 0 if you assume no one rated it
df['Rating'].fillna(0, inplace=True)

# Drop rows where Price or Capacity is missing (essential features)
df.dropna(subset=['Price_Rupees', 'Capacity_Litres'], inplace=True)

# Optionally: Fill other missing values with reasonable defaults
df['Discount_Percent'].fillna(0, inplace=True)
df['Compressor_Warranty_Years'].fillna(df['Compressor_Warranty_Years'].median(), inplace=True)
df['Review_And_Ratings'].fillna("No reviews", inplace=True)
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
# Remove ₹ and commas from Price
df['Price'] = df['Price'].str.replace('₹', '', regex=True).str.replace(',', '', regex=True).astype(float)

# Convert Discount to numeric
df['Discount'] = df['Discount'].astype(float)

# Convert Capacity to numeric
df['Capacity'] = df['Capacity'].astype(float)

# Convert Rating to numeric
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Convert Compressor Warranty to numeric
df['Compressor_Warranty_Years'] = pd.to_numeric(df['Compressor_Warranty_Years'], errors='coerce')
```

```python
df[['Total_Ratings', 'Total_Reviews']] = df['Review_rating'].str.extract(r'([\d,]+)\s+Ratings\s+&\s+([\d,]+)\s+Reviews')
df['Total_Ratings'] = df['Total_Ratings'].str.replace(',', '', regex=True).astype(float)
df['Total_Reviews'] = df['Total_Reviews'].str.replace(',', '', regex=True).astype(float)
```

```python
# Fill numeric missing values with median
numeric_cols = ['Rating', 'Capacity', 'Discount', 'Price', 'Compressor_Warranty_Years', 'Total_Ratings', 'Total_Reviews']
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill categorical missing values with mode
df['title'] = df['title'].fillna("Unknown")
df['Brand'] = df['Brand'].fillna(df['Brand'].mode()[0])
```

```python
# Drop the original review text column
df.drop(columns=['Review_rating'], inplace=True)
```

```python
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]
```
